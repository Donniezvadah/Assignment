## <font color=green>Exercise 07</font> <a class = "anchor" id="exercise-07"></a> 
(*) Write a Python function (a function using **def**) in _SageMath_ to list enough numbers for a general pair $\{a, b\}$ where $a$ and $b$ are natural numbers to guess the conductor of $a$ and $b$. (And explain it!)

```python
def guesswork(a, b):
    if gcd(a,b) != 1:
        return f" The Frobenius and Conducter do not exist for ({a}, {b}) given."
    else:
        upper_bound = 10000
        consecutive_count = b - a +100
        found_values = set()
        consecutive_values = set()

        for x in range(upper_bound):
            for y in range(upper_bound):
                n = a * x + b * y
                found_values.add(n)

                # Check for consecutive values
                current_consecutive = {n + i for i in range(consecutive_count)}
                if current_consecutive.issubset(found_values):
                    consecutive_values = current_consecutive
                    break
            if consecutive_values:
                break

    return sorted(consecutive_values)

```

### Rules and Documentation of the function

**Here, the function is trying to guess a frobenius number given the following**
* A list of numbers will be created. The number we can make up to the certain stated stopping rule
* The Code stops running when the stopping rule is observed
* The rule is we need to generate 100+ (b-a) numbers that are consecutive ( Subject to changes on this)
* The numbers generated above will show that there is now a bit of consistence in the list of Natural Numbers
* The functions has an upper bound of searching the numbers that are consecutive
* The numbers are added to the set (Ascending order)
* The $\mathbb{F}$ : Frobenius number guess is that it is the 1st entry in the set of consecutive numbers
* The code is limited on the search boundary as it will create more and more values of $n$ during the search process hence the code will we a bit slow for higher values of upper_boound
